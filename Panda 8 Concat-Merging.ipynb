```python
import pandas as pd

Pakistan_weather = pd.DataFrame({
    "city":["karachi","lahore","islamabad"],
    "temperature":[42,28,30],
    "humidity":[80,50,60]
})
Pakistan_weather
```

```python
import pandas as pd

usa_weather = pd.DataFrame({
    "city":["newyork","chicago","orlando"],
    "temperature":[32,22,35],
    "humidity":[66,57,48]
})
usa_weather
```

```python
df = pd.concat([Pakistan_weather, usa_weather])
df
```

```python
df = pd.concat([Pakistan_weather, usa_weather], ignore_index=True)
df
```

```python
df = pd.concat([Pakistan_weather, usa_weather], keys=["Pakistan","USA"])
df
```
